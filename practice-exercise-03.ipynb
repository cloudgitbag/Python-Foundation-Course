### Question 9:

Reverse the integer given below.

n = 5623

#### Expected Output:

3265

```python
n = 5623
s=str(n)
sr=s[::-1]
print(sr)

############
rev = 0

while n > 0:
    rem = n % 10
    rev = (rev * 10) + rem
    n = n // 10
    
print(rev)
```
